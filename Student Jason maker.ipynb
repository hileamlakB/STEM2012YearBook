```python
studentList = open("Student list.txt",mode = "r").read().split("\n")
```

```python
jsonFile = []
```

```python
{
    "fullName":"",
    "images":["","","","","",""],
    "cartoonImage":"",
    "friendNotes": ["",""],
    "theam":[]
}
```

```python
for eachStudent in studentList:
    data = eachStudent.split(",")
    fullName = data[0]
    ids = fullName.split(" ")[0][0:5]+fullName.split(" ")[1][0:2]
    images = [ids+str(i+1)+".jpg" for i in range(6)]
    cartoonImage = ids+"cartoonImage.jpg"
    friendNotes = [ids+"Note"+str(i+1)+".jpg" for i in range(2)]
    theam = data[1]
    
    fullInfo = {
        "fullName":fullName,
        "images":images,
        "cartoonImage":cartoonImage,
        "friendNotes": friendNotes,
        "theam":theam
        }
    jsonFile.append((ids,fullInfo))

    
```

```python
jsonFile
```

```python
for eachStudent in jsonFile:
    print("'",eachStudent[0],"':",eachStudent[1],",")
```
